```python
import re

def remove_non_alphanumeric(string):
    return re.sub(r'[^\w]', '', string)

input_string = "Hello, world! 123"
output_string = remove_non_alphanumeric(input_string)
print(output_string)
```

```python
import re
import urllib.request

url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]

def capitalize_iris_species(string):
    return re.sub(r'\b(\w)', lambda x: x.group(1).upper(), string)

strip_iris = re.compile(r'([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),Iris-([a-z]+)')
iris = [strip_iris.sub(lambda x: f"{capitalize_iris_species(x.group(5))},{x.group(4)},{x.group(3)},{x.group(2)},{x.group(1)}", line) for line in iris]

iris
```
