### Give 3 real world examples of Cognitive Bias
***

Cognitive bias is defined as a subconscious error in thinking that leads you to misinterpret information from the world around you, and affects the rationality and accuracy of decisions and judgments. Biases are unconscious and automatic processes designed to make decision-making quicker and more efficient.
Cognitive biases can be caused by a number of different things, such as heuristics (mental shortcuts), social pressures, and emotions.

Simply put, cognitive bias occurs when people allow their emotions, beliefs or personal preferences to influence their judgements, which could lead to inaccurate decisions.

According to the Cognitive bias codex (below) there are an estimated 180 (and counting) types of cognitive bias

## Cognitive bias examples

### Dunning Kruger effect
***

The Dunning-Kruger effect is a bias where people overestimate their knowledge or ability in a given topic. This is quite common in todays very online world, social media is filled with people who believe themselves to be experts in a given field despite not having the knowledge or experience to back it up, this can lead to poor outcomes
